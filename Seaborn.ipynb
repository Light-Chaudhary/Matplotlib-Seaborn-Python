```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
print(sns.get_dataset_names())
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins = penguins.dropna()
```

```python
penguins = penguins.rename(columns = {"bill_length_mm": "bill_length", "bill_depth_mm": "bill_depth", "flipper_length_mm": "flipper_length", "body_mass_g": "body_mass"})
```

```python
sns.scatterplot(x = "bill_depth", y = "bill_length", hue = "species", style = "sex", size = "body_mass", data = penguins)
plt.title("Bill Length vs Bill Depth ")
plt.legend(loc = "best")
plt.xlabel ("Bill Depth (mm)")
plt.ylabel("Bill Length (mm)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
