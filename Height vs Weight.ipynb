```python
import matplotlib.pyplot as plt
import numpy as np

heights = [150, 155, 160, 165, 170, 175, 180, 185, 190]
weights = [45, 50, 55, 60, 65, 70, 75, 80, 85]

heights_normalized = np.array(heights)

plt.style.use("petroff10")
plt.scatter(weights, heights, c = heights_normalized, label = "Height vs Weight",cmap= "viridis")
plt.legend(loc = "best")
plt.colorbar(label = "heights")
plt.xlabel ("Weights")
plt.ylabel("Heights")
plt.title("Height vs Weight Comparison")
plt.grid(True)
plt.tight_layout()
plt.show()
```
