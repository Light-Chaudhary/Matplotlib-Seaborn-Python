```python
import matplotlib.pyplot as plt 

group1_heights = [150, 155, 160, 165, 170]
group1_weights = [45, 50, 55, 60, 65]

group2_heights = [160, 165, 170, 175, 180]
group2_weights = [50, 55, 60, 70, 75]

group3_heights = [170, 175, 180, 185, 190]
group3_weights = [60, 65, 70, 80, 85]

plt.subplot(1,3,1)
plt.scatter(group1_weights, group1_heights, color = "red")
plt.title("Group 1")
plt.grid(True)
plt.xlabel("Weights in kg")
plt.ylabel("Heights in cm")

plt.subplot(1,3,2)
plt.scatter(group2_weights, group2_heights, color = "blue")
plt.title("Group 2")
plt.xlabel("Weights in kg")
plt.ylabel("Heights in cm")
plt.grid(True)

plt.subplot(1,3,3)
plt.scatter(group3_weights, group3_heights, color = "green")
plt.title("Group 3")
plt.xlabel("Weights in kg")
plt.ylabel("Heights in cm")
plt.grid(True)

plt.tight_layout()
plt.suptitle("Comparison of heights and weight across various groups", fontsize = 16,y = 1.05)
plt.show()
```
